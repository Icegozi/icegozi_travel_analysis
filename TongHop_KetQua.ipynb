# Tổng hợp xu hướng quan tâm và hành vi đánh giá

Notebook chạy sau khi đã có `destination_season_stats.csv` và phản hồi khảo sát hợp lệ.

```python
from pathlib import Path
import pandas as pd

project_root = Path.cwd()
if not (project_root / 'data').exists():
    project_root = project_root.parent
analysis_dir = project_root / 'data' / 'analysis'
interest = pd.read_csv(analysis_dir / 'destination_season_stats.csv')
rating_file = analysis_dir / 'destination_rating_stats.csv'

if not rating_file.exists():
    print('Chưa có dữ liệu khảo sát. Hãy chạy KHAO_SAT/PhanTich_KhaoSat.ipynb sau khi thu thập phản hồi.')
else:
    ratings = pd.read_csv(rating_file)
    summary = interest.groupby('destination', as_index=False).agg(
        article_count=('article_count', 'sum'),
        total_views=('total_views', 'sum'),
        median_views_per_day=('median_views_per_day', 'median'))
    summary = summary.merge(ratings, on='destination', how='outer')
    summary.to_csv(analysis_dir / 'destination_combined_summary.csv', encoding='utf-8-sig', index=False)
    display(summary.sort_values('median_views_per_day', ascending=False))
```
